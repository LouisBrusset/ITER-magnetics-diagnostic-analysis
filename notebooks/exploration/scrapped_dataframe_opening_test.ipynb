```python
import xarray as xr
import numpy as np
from pathlib import Path
```

```python
# Path to the NetCDF file
DATA_PATH = Path().absolute().parent.parent / "data/raw/mscred/data_magnetics_mscred.nc"
```

```python
ds = xr.open_dataset(DATA_PATH)
ds.load()
```

```python
for var in ds.data_vars:
    print(f"Variable: {var}")
    print(ds[var].shape)
    print("===============")
```

```python
ds["b_field_pol_probe_omv_voltage"].shape
```

```python
# Count total channels
total_channel_number = 0
for var in ds.data_vars:
    if var not in ["time", "shot_index"]:
        try:
            total_channel_number += ds[var].shape[1]
            print(f"Variable '{var}' has {ds[var].shape[1]} channels.")
        except IndexError:
            total_channel_number += 1
            print(f"Variable '{var}' has only 1 channel.")
print(f"\nTotal number of channels: {total_channel_number}")
```

```python
ds.close()
```
